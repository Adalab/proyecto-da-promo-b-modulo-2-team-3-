```python
url = "https://dev.adalab.es/api/cinema/movies?year=2010&genre=Drama"

genreList = ["Romance", "Thriller"]
movieList = []  #Lista que almacenará las tuplas con la info de las películas

for genre in tqdm(genreList): 
       for year in range(2009,2025):   #Recordatorio: range excluye el último valor proporcionado
        sleep(30)    #Incluimos un tiempo muerto para no saturar el servidor con peticiones
        
        url = f"https://dev.adalab.es/api/cinema/movies?year={year}&genre={genre}"      #Incluimos el endpoint genérico
        response = requests.get(url)    #Realizamos la petición a la API
               
        if response.status_code == 200:     ## Verificamos el resultado de la petición
            print(f'Response for your request for genre {genre} and year {year} is OK')
        else:
            print(f'There is a problem with your request for genre {genre}, year{year}. La razón del fallo es {response.reason}')    

        movies = response.json()["results"]     # Almacenamos en la variable movies el json devuelto por la API
        
        for movie in movies:    #Para cada elemento de results del json devuelto 
            
            if movie["type"] == "Movie" or movie["type"] == "Short": ## Comprobamos si es una "Movie" o es un "Short", verifico en el campo type
                
                film_list = []                      #Cambiado el nombre de la lista de 'resultado' a 'film_list' por estar todas las variables en inglés
                film_list.append(movie["type"])     #Añadimos a la lista film_list toda la información del json que nos interesa
                film_list.append(movie["title"])
                film_list.append(year) #No entiendo que añada el año y el género a la lista sin escribir append.(movie['year'])
                film_list.append(genre)
                film_list.append(movie["id"])
                film_list.append(movie["idOwner"])  #Comprobar en la grabación de clase que este no hace falta. En las instrucciones del proyecto solo dice id de la película
                movieList.append(tuple(film_list))
                
            else:
                pass

print(movieList);

# Falta la parte de index y enumerate en el bucle for para que añada un número a cada película
#Intentar programar con funciones y con clases
```

```python
import pandas as pd

# Lista de archivos CSV a unir (rutas relativas)
archivos_csv = ["data/file1.csv", "data/file2.csv", "data/file3.csv"]

# Lista para almacenar los DataFrames
dataframes = []

# Leer cada archivo CSV y añadir el DataFrame a la lista
for archivo in archivos_csv:
    df = pd.read_csv(archivo)
    dataframes.append(df)

# Concatenar todos los DataFrames en uno solo
df_concatenado = pd.concat(dataframes, ignore_index=True)

# Guardar el DataFrame concatenado en un archivo CSV
df_concatenado.to_csv("data/archivo_unido.csv", index=False)

```
